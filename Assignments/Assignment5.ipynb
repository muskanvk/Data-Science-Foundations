```python
#Create a python program that should take the input “filename” in .csv format from user and in case of duplicate entries, consider the last entries only. For missing values, you need to put mean of value column.

import pandas as pd
import numpy as np

fileinput = str(input("Which file do you want? "))
if not ".csv" in fileinput:
    fileinput += ".csv"
    
df = pd.read_csv(fileinput)
new = df.drop_duplicates(subset=['No'],keep='last')
new['Value'] = pd.to_numeric(new['Value'], errors ='coerce') #handling string values
new = new.fillna(new.mean())
new
```

```python
import pandas as pd
from functools import reduce
df1 = pd.read_csv("File1.csv")
df1 = df1.drop_duplicates(subset=['No'],keep ='last')
df1['Value1'] = pd.to_numeric(df1['Value1'], errors ='coerce')


df2 = pd.read_csv("File2.csv")
df2 = df2.drop_duplicates(subset=['No'],keep ='last')
df2['Value2'] = pd.to_numeric(df2['Value2'], errors ='coerce')

df3 = pd.read_csv("File3.csv")
df3 = df3.drop_duplicates(subset=['No'],keep ='last')
df3['Value3'] = pd.to_numeric(df3['Value3'], errors ='coerce')

dfs = [df1, df2, df3]
df_final = reduce(lambda left,right: pd.merge(left,right,on='No',how='outer'), dfs)
pd.DataFrame.to_csv(df_final, 'merged.csv', index=False)

new = pd.read_csv("merged.csv")
new2 = new1.sort_values(by=["No"])
new2['Value1']=new2['Value1'].fillna(new['Value1'].min())
new2['Value2']=new2['Value2'].fillna(new['Value2'].min())
new2['Value3']=new2['Value3'].fillna(new['Value3'].min())
new2
```
